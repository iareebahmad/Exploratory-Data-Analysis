Plotting Visuals

```python
df.head(5)
```

```python
# Line Plot
plt.plot(df['Country'],df['Revenue'])
plt.xlabel("Country")
plt.ylabel("Revenue Generated")
plt.title("Revenue Index 2025")
plt.show()
```

```python
# Bar Graph
plt.bar(df['Country'],df['Revenue'])
plt.xlabel("Country")
plt.ylabel("Revenue Generated")
plt.title("Revenue Index 2025")
plt.show()
```

```python
# Pie Chart
plt.pie(df['Revenue'],labels = df['Country'],autopct='%1.1f%%')
plt.title("Revenue Share by Country")
plt.show()
```

```python
# Box Plot
plt.boxplot(df['Revenue'])
plt.ylabel("Revenue of countries")
plt.title("Identifying Outlier")
plt.show()
```

```python
# Histogram
plt.hist(df['Revenue'], bins = 6, color = 'green', edgecolor = 'black')
plt.title("Revenue histogram")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df['Revenue'],df['Country'], color = 'red')
plt.title("Simple Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
