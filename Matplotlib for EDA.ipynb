```python
import matplotlib.pyplot as plt
```

# plt.VisualName
1. Normal Graph - plt.plot(x,y)
2. Bar Graph - plt.bar(x,y)
3. Pie Chart - plt.pie(sizes, labels = macros)
4. BoxPlot - plt.boxplot(data) -- For Median and Outliers 
5. Histogram - plt.hist(data, bins=6,color = "skyblue", edgecolor="black")  -- For frequency
6. Scatter Plot - plt.scatter(x, y, color="green") -- Trend Detection and Spot Outliers

```python
# Graphs
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("Independant")
plt.ylabel("Dependant")
plt.title("Sq of Numbers")
plt.show()
```

```python
# bar graph
x = ["India","Australia","USA","Canada"]
y = [20,19,23,14]

plt.bar(x,y)
plt.xlabel("Country")
plt.ylabel("Revenue in Lacs")
plt.title("Development Index")
plt.show()
```

```python
# Pie Chart
sizes = [30,30,25,10,5]  # Should sum upto 100
macros = ["Carbs","Fats","Protein","Roughage","Water"]

plt.pie(sizes, labels = macros)
plt.title("Meal Portion")
plt.show()
```

```python
# BoxPlot - For Outliers
data = [7, 8, 5, 6, 9, 12, 15, 14, 10, 8, 7, 6, 5, 25]

# Boxplot
plt.boxplot(data)
plt.title("Boxplot")
plt.ylabel("Values")
plt.show()
```

 This will show:

1. The median (middle line inside the box)

2. The quartiles (box edges)

3. The whiskers (spread of data)

4. Any outliers (points outside the whiskers, e.g., 25 here)

```python
# Histogram
data = [7, 8, 5, 6, 9, 12, 15, 14, 10, 8, 7, 6, 5, 25]

plt.hist(data, bins=6,color = "skyblue", edgecolor="black")
plt.title("Sample histogram")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.show()
```

 This shows how values are distributed into bins (groups).
For example, it will count how many numbers fall between ranges (like 58, 912, etc.).

```python
# Scatter Plot
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="green")
plt.title("Simple Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

1. Spot outliers

        If one point lies far away from the rest, its an unusual observation.

2. Trend detection

        Even without a formal line of best fit, you can see if data tends to increase or decrease.
